```python
%reload_ext autoreload
%autoreload 2

import pandas as pd
from google_trends_analyzer import AITrendAnalyzer
from pytrends.request import TrendReq
import pandas as pd
import numpy as np
from typing import List, Dict, Tuple
from datetime import datetime, timedelta
import os
```

**This implementation:**
1. Creates a comprehensive trend score based on multiple indicators:
    - Momentum (20% weight): Compares recent activity to previous period
    - Growth Trend (35% weight): Uses linear regression to determine overall direction
    - Peak Ratio (35% weight): Measures current popularity relative to peak
    - Volatility (10% negative weight): Penalizes unstable patterns
2. Provides a confidence score based on the consistency of different metrics.
3. Determines trending status using both the trend score and confidence level.

```python
analyzer = AITrendAnalyzer()

# Example AI tools to analyze
ai_tools = [
    "DALL-E 2",
    "Notion AI",
    "GAMMA AI",
    "NVIDIA NIM",
    "Meta Llama 3.2",
    "OpenAI (o1)",
    "Perplexity AI",
    "Claude",
    "Google Gemini",
    "VectorShift",
    "BotPress",
    "Replit AI",
    "Cursor AI",
    "Denser AI",
    "Unstract",
    "AIAC",
    "PearAI",
    "NotebookLM",
    "Aider",
    "V0 AI",
    "Melty"
]

# Analyze each tool
results = {}
for tool in ai_tools:
    results[tool] = analyzer.calculate_trend_score(tool)

# Create DataFrame
df = analyzer.create_results_dataframe(results)

# Print results to console
print("\nAI Tools Trend Analysis:")
print("-" * 50)
print(df)

# Export to Excel
output_dir = "trend_analysis_results"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
filename = os.path.join(output_dir, f"ai_tools_trend_analysis_{timestamp}.xlsx")

exported_file = analyzer.export_to_excel(df, filename)
print(f"\nResults exported to: {exported_file}")
```
